#### 股票殖利率＝現金股利 / 股價 (越高越好)
#### 本益比(PER) = 每股市價 / 每股盈餘(EPS) （越低越好）
#### 股價淨值比(PBR) = 股票市值 / 每股淨值 （股價淨值比小於1時，代表現在比較便宜；股價淨值比大於1時，代表現在比較昂貴）

```python
import requests
import json
import pandas as pd
```

```python
res = requests.get('http://www.tse.com.tw/exchangeReport/BWIBBU_d?response=json&date=20180326&selectType=ALL')
print(res.text)
```

```python
jres = json.loads(res.text)
jres
```

```python
jres['stat']
```

```python
jres['data']
```

```python
pd.set_option('display.max.rows',913)
df_temp = pd.DataFrame(jres['data'],columns=jres['fields'])
df_temp
```

## 預處理

```python
df = df_temp[['證券代號','證券名稱','殖利率(%)','本益比','股價淨值比']]
df = df[(df['殖利率(%)']!='-') & (df['本益比']!='-') & (df['股價淨值比']!='-')].reset_index(drop=True)
```

```python
for i in range(len(df)):
    df.loc[i,'殖利率(%)'] = df.loc[i,'殖利率(%)'].replace(',','')
    df.loc[i,'本益比'] = df.loc[i,'本益比'].replace(',','')
    df.loc[i,'股價淨值比'] = df.loc[i,'股價淨值比'].replace(',','')
```

```python
df[['殖利率(%)','本益比','股價淨值比']] = df[['殖利率(%)','本益比','股價淨值比']].astype(float)
df
```

```python
df.sort_values(['殖利率(%)', '本益比','股價淨值比'], ascending=[0, 1, 1])
```

```python
df
```

```python
X = df[['殖利率(%)','本益比']]
```

# Kmeans

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X_std = sc.transform(X)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(X_std)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.family']='SimHei' #顯示中文('SimHei' for MacOS)

plt.figure(figsize=(10,8))
plt.scatter(X['殖利率(%)'],X['本益比'],c=y_pred)
plt.xlabel('殖利率(%)', fontsize=20)
plt.ylabel('本益比', fontsize=20)
```

```python
X = df[['殖利率(%)','股價淨值比','本益比']]
```

```python
sc = StandardScaler()
sc.fit(X)
X_std = sc.transform(X)
```

```python
km = KMeans(n_clusters=5)
y_pred = km.fit_predict(X_std)
```

```python
# 回復標準化前原值
sc.inverse_transform(km.cluster_centers_)
```

```python
for i in km.cluster_centers_:
    print(i)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = Axes3D(fig)
ax.scatter(xs=X['殖利率(%)'],ys=X['股價淨值比'],zs=X['本益比'],c=y_pred)
ax.set_xlabel('殖利率(%)',fontsize=18)
ax.set_ylabel('股價淨值比',fontsize=18)
ax.set_zlabel('本益比',fontsize=18)

c = 0
for i in sc.inverse_transform(km.cluster_centers_):
    ax.text(i[0], i[1], i[2], c, color='black', fontsize=20)
    c += 1
```
