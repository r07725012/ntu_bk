## 1. X軸雙變項作圖（折線圖、長條圖） 

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# 調整作圖大小
fig = plt.figure(figsize=(4,3))

x = [1,2,3,4,5]
y = [123,432,823,192,12]
x2 = [1,2,3,4,5]
y2 = [432,12,432,847,223]

plt.plot(x,y,'-')
plt.plot(x2,y2,'-')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')

fig = plt.figure(figsize=(4,3))

x = [1,2,3,4,5]
y = [123,432,823,192,12]
x2 = [1,2,3,4,5]
y2 = [432,12,432,847,223]

plt.bar(np.array(x)-0.2,y,color = 'g',width=0.4,align='center')
plt.bar(np.array(x2)+0.2,y2,color = 'b',width=0.4,align='center')
plt.show()
```

## 3. 標題、X、Y軸名稱（中文處理與字體選擇）

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig = plt.figure(figsize=(4,3))
x = [1,2,3,4,5]
y = [123,432,823,192,12]

plt.rcParams['font.family']='DFKai-SB' #顯示中文 (for Win10)

plt.bar(x,y,align='center')
plt.title('中文標題') #標題名稱
plt.xlabel('X軸') #X軸名稱
plt.ylabel('Y軸') #Y軸名稱
plt.xticks(x, ['刻度A','刻度B','刻度C','刻度D','刻度E']) #X軸刻度名稱

plt.show()
```

```python
import matplotlib
[f.name for f in matplotlib.font_manager.fontManager.ttflist]
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig = plt.figure(figsize=(4,3))
x = [1,2,3,4,5]
y = [123,432,823,192,12]

font1 = {'fontname':'Times New Roman'}
font2 = {'fontname':'Arial'}

plt.bar(x,y,align='center')
plt.title('Title',**font1) #標題名稱
plt.xlabel('X bar',**font1) #X軸名稱
plt.ylabel('Y bar',**font1) #Y軸名稱
plt.xticks(x, ['tick A','tick B','tick C','tick D','tick E'],**font2) #X軸刻度名稱

plt.show()
```

## 4. 多圖合一

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plt.figure(figsize=(6,6))
x = [1,2,3,4,5]
y = [123,432,823,192,12]

ax1 = plt.subplot(211) #2x1的第1張圖
plt.plot(x,y,'-')
plt.setp(ax1.get_xticklabels(), visible=False) #不顯示x軸ticks
plt.subplot(212,sharex=ax1) #2x1的第2張圖
plt.bar(x,y,align='center')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,'-')
plt.bar(x,y,align='center')
plt.show()
```

## 5. 折線圖種類

##### color + style

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig = plt.figure(figsize=(4,3))
x = [1,2,3,4,5]
y = [123,432,823,192,12]
plt.plot(x,y,'-')
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
#plt.plot(x,y,'g--') #green
plt.plot(x,y,color='g',linestyle='--')
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(x,y,'b-.') #blue
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(x,y,'b-o') #blue
plt.show()
```

```python
# 不連線的折線圖 => 散佈圖
fig = plt.figure(figsize=(4,3))
plt.plot(x,y,'*') #blue
plt.show()
```
